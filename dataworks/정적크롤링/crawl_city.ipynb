```python
import requests 
from bs4 import BeautifulSoup

url = "https://www.seoul.go.kr/main/index.jsp"
response = requests.get(url)
print(response.text)

soup = BeautifulSoup(response.text, 'html.parser')
#print(soup)
first_li = soup.select_one("li.public")
print(first_li)
print(first_li.text)



```
