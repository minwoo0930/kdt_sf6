### 시각화 라이브러리 사용 -Matplotlib

```python
import matplotlib.pyplot as plt 

plt.title("Basic Graph")  # 그래프 이름 출력 
#plt.plot([1, 2, 3, 4]) # 값이 하나만 입력되면 Y축 값이 입력된다. 
plt.plot([1,2,3,4],[10, 40, 20, 30])# plt(가로(x)축 값, 세로(Y)축 값)
plt.show() # 시각화 출력 

```

```python
import matplotlib.pyplot as plt 

plt.rc('font', family = 'Malgun Gothic')
plt.title("시험 성적 그래프")# 그래프 이름 출력   
#plt.plot([1, 2, 3, 4]) # 값이 하나만 입력되면 Y축 값이 입력된다. 
plt.plot([1,2,3,4],[90, 80, 85, 70])# plt(가로(x)축 값, 세로(Y)축 값)
plt.xlabel("학생번호")# 한국어로 하면 안됨 -> plt.rc 함수로 한글 글꼴 명시해야함/ x축 제목 기입 함수 
plt.ylabel("시험점수")# y축 제목 기입 함수 
plt.show() # 시각화 출력 
```

```python
import matplotlib.pyplot as plt 

plt.rc('font', family = 'Malgun Gothic')
plt.title("범례 및 색상 표시")# 그래프 이름 출력   
plt.plot([10,20,30,40], label = "circle", color= 'r') #범례 표시(label = ""), 컬러 표시(color = "")
plt.plot([40,30,20,10], label = "triangle", color= 'g') 
plt.legend() #범례(소제목)를 표시할 때 반드시 필요한 함수 -> 없으면 범례 표시 안됨 
plt.show() # 시각화 출력 
```

```python
import matplotlib.pyplot as plt 

plt.rc('font', family = 'Malgun Gothic')
plt.title("범례 및 색상 표시")# 그래프 이름 출력   
plt.plot([10,20,30,40], 'b.', label = "circle") # 'b.' -> 파란색 동그라미 마커를 의미 
plt.plot([40,30,20,10], 'r^', label = "triangle") #'r^' -> 빨간색 삼각형 마커를 의미 
plt.legend() #범례(소제목)를 표시할 때 반드시 필요한 함수 -> 없으면 범례 표시 안됨 
plt.show() # 시각화 출력 
```
